## AudioLDM

```python
from diffusers import AudioLDMPipeline, AudioLDM2Pipeline
import torch
import scipy
import os
from IPython.display import Audio

audio_folder = '/nas/home/fronchini/urban-sound-class/audio'
os.makedirs(audio_folder, exist_ok=True)
model = "audioldm-s-full-v2"
repo_id = f"cvssp/{model}"
pipe = AudioLDMPipeline.from_pretrained(repo_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
class_sound = 'dog'
prompt = f"{class_sound} barking"
audio = pipe(prompt, num_inference_steps=100, audio_length_in_s=5.0).audios[0]
#audio = pipe(prompt, num_inference_steps=100, audio_length_in_s=5.0)

audio_file_out = f"{class_sound}_{model}.wav"
audio_file_out_path = os.path.join(audio_folder, audio_file_out)
scipy.io.wavfile.write(audio_file_out_path, rate=16000, data=audio)
```

```python
print(f"Audio from model: {model}")
Audio(audio, rate=16000)
```
